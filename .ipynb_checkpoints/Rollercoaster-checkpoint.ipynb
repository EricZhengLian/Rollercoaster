From the info report, we noticed that the columns "Drop(feet)", "G Force", "Vertical Angle (degrees)" and "CoasterBuzz Ranking" have fewer than half of the valuable data. So we decide to get rid of these columns.

```python
rc_ranking = rc.drop(columns = ['Drop (feet)', 'G Force', 'Vertical Angle (degrees)'])
print(rc_ranking.info())
rc = rc_ranking.drop(columns = "CoasterBuzz Ranking")
print(rc.info())
```

We want to fill in the NaN values with the mean of their corresponding columns.

```python
#rc['Duration (min:sec)'] = rc['Duration (min:sec)'].replace(r'^\s+$', np.nan, regex=True)
#rc['Duration (min:sec)'] = pd.to_datetime(rc['Duration (min:sec)'])
#rc['Duration (min:sec)'].min
```

```python
rc['Speed (mph)'].fillna((rc['Speed (mph)'].mean()),inplace = True)
rc['Length (feet)'].fillna((rc['Length (feet)'].mean()),inplace = True)
rc['Height (feet)'] = rc['Height (feet)'].astype(float)
rc['Height (feet)'].fillna((rc['Height (feet)'].mean()),inplace = True)
#rc[rc['Duration (min:sec)'].isnull() == False]['Duration (min:sec)'] = pd.to_datetime(rc[rc['Duration (min:sec)'].isnull() == False]['Duration (min:sec)'])


#rc['Duration (min:sec)'] = rc['Duration (min:sec)'].replace(r'^\s+$', np.nan, regex=True)

for m in range(len(rc)):
    try:
        value = rc['Duration (min:sec)'][m].split(':')
        i = 60
        total_sec = 0
        for numbers in value:
            total_sec += i*int(numbers)
            i = 1
        rc.loc[m, 'Duration (min:sec)'] = int(total_sec)
    except:
        pass

rc.info()








    























```
